```python
import matplotlib.pyplot as plt
data = [2, 3, 4, 5, 6]
plt.plot(data)
plt.show()
```

```python
import matplotlib.pyplot as plt
price = [200, 300, 400, 500, 600]
count = [31,29,25,38,26]
plt.plot(price,count,marker="o")
plt.xlabel("price")
plt.xlabel("count")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-4,4,0.01)
y = x**2+1
plt.plot(x,y)
plt.title("y=x**2+1")
plt.grid(True)
plt.show()
```
